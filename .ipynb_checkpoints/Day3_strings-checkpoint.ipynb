String Traversal

```python
string=input("Enter a String : ")
pos=0
for i in string:
    print(i," at position ",pos)
    pos+=1
```

String Length Without len()

```python
string=input("Enter string : ")
count=0
for i in string:
    count+=1
print("String length is ",count)
```

vowels/consonent logic

```python
string = input("Enter a String : ")
vowels="aeiou"
vow_count=0
cons_count=0
for i in string:
    if i in vowels:
        vow_count+=1
    elif i!=" ":
        cons_count+=1
print("Vowels = ",vow_count," Consonent = ",cons_count)
```

Reverse String 2 ways

```python
string=input("Enter String : ")
revstr=""
for i in range(len(string)-1,-1,-1):
    revstr+=string[i]
print(revstr)
    
```

With Slicing

```python
string=input("Enter String : ")
revstr=string[::-1]
print(revstr)
```

Palindrome String

```python
string=input("Enter String to check Palindrome or not : ")
revstr=""
for i in range(len(string)-1,-1,-1):
    revstr+=string[i]
if string==revstr:
    print("Yes Palindrome")
else:
    print("Not a Palindrome")
```

Firt non Repeting Character

```python
string=input("Enter string : ")
for i in string:
    if string.count(i)==1:
        print(i)
        break
```

```python
string1=input("Enter string 1 : ")
string2=input("Enter string 2 : ")
if len(string1) != len(string2):
    print("False")
else:
    flag=False
    for i in string1:
        if string1.count(i) != string2.count(i):
            print("False")
            flag=True
            break
    if flag==False:
        print("True")
```

count words Without Spit()

```python
string = input("Enter the string : ")
inword=False
count=0
for i in string:
    if i != " " and not inword:
        count+=1
        inword=True
    elif i==" ":
        inword=False
print(count)
```

```python
string=input("Enter String : ")
print(len(string.split()))
```

Longest Word in Sentence

```python
string = input("Enter the String : ")
l1=string.split()
largest=0
for i in l1:
    if len(i)>largest:
        largest=len(i)
        s = i
print(s)
```

remove duplicates in word 

```python
string=input("Enter string : ")
newstr=""
unique=""
for i in string:
    if i not in unique:
        unique+=i
        newstr+=i
print(newstr)
```

most occuring character

```python
string=input("Enter String : ")
unique=""
for i in string:
    if i not in unique:
        unique+=i
large=0
for i in unique:
    if string.count(i)>large:
        large=string.count(i)
for i in unique:
    if string.count(i)==large:
        print(i,large)
    
```
