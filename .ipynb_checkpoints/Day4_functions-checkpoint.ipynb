Function to Check Prime 

```python
def prime(num):
    flag=False
    for i in range(2,num):
        if num%i==0:
            flag=True
            break
    if flag==False:
        print("Prime")
    else:
        print("not Prime")

num=int(input("Enter Number : "))
prime(num)
```

Function to return Factorial

```python
def fact(num):
    fact=1
    for i in range(1,num+1):
        fact*=i
    return fact
num=int(input("Enter Number : "))
print(fact(num))
```

Function to count digits

```python
def count_digit(num):
    c=0
    while num>0:
        digit=num%10
        c+=1
        num//=10
    return(c)
num=int(input("Enter Number : "))
print(count_digit(num))
```

Function to reverse the number

```python
def rev_num(num):
    rev=0
    while num>0:
        digit=num%10
        rev=rev*10+digit
        num//=10
    return(rev)
num=int(input("Enter Number : "))
print(rev_num(num))
```

Sum of digits using Recursion

```python
def sumofdigit(num):
    if num==0:
        return 0
    return num%10+sumofdigit(num//10)
num=int(input("Enter Number : "))
print(sumofdigit(num))
```

Fibonacci using Recursion

```python
def fibo(num):
    if num==0:
        return 0
    if num==1:
        return 1
    return fibo(num-1)+fibo(num-2)
num=int(input("Enter Number : "))
print(fibo(num))
```

GCD using Recursion

```python
def gcd(a,b):
    if b==0:
        return a
    return gcd(b,a%b)
num1=int(input("Enter Number 1 : "))
num2=int(input("Enter Number 2 : "))
print(gcd(num1,num2))
```

String Palindrome using Recursion

```python
def palindrome(string,i=0):
    if i>=len(string)//2:
        return True
    if string[i]!=string[-i-1]:
        return False
    return palindrome(string,i+1)

string=input("Enter String to check palindrome : ")
print(palindrome(string))
```

Count vowels in String with function

```python
def count_vow(string,c=0):
    vow="aeiou"
    for i in string:
        if i in vow:
            c+=1
    return c
string=input("Enter string : ")
print(count_vow(string))
```

Power of Number using recursion

```python
def power(a,b,p=1):
    if b==0:
        return p
    return power(a,b-1,p*a)
print(power(3,4))
    
```

```python
def reverse(num,rev=0):
    if num==0:
        return rev
    return reverse(num//10,rev * 10 + num % 10)
num=int(input("Enter Number : "))
print(reverse(num))
```

LCM using GCD using recursion

```python
#lcm formula is lcm(a,b)=(a*b)//gcd(a,b)
def gcd(a,b):
    if b==0:
        return a
    return gcd(b,a % b)
num1 = int(input("Enter num1 : "))
num2 = int(input("Enter num2 : "))
print((num1*num2)//gcd(num1,num2))
```

Strong Number using recursion

```python
def fact(num):
    if num==0 or num==1:
        return 1
    return num*fact(num-1)
def strong(num):
    if num==0:
        return 0
    return fact(num%10)+strong(num//10)
num = int(input("Enter Number"))
if strong(num)==num:
    print("strong")
else:
    print("not Strong")
```

Perfect Number using recursion

```python
def perfect(num,i=1,s=0):
    if i==num:
        return s
    if num % i ==0:
        s+=i
    return perfect(num,i+1,s) 
num=int(input("Enter number : "))
print("Perfect") if num==perfect(num) else print("not Perfect")
```

Find sum of digits until single digit using recursion

```python
def sumdigit(num,s=0):
    if num==0:
        return s
    return sumdigit(num//10,s+num%10)
def digit_root(num):
        if num<10:
            return num
        return digit_root(sumdigit(num))
num=int(input("Enter Number : "))
print(digit_root(num))
```

Fibonacci using recursion for N of series

```python
def fibo(num):
    if num==0 :
        return 0
    if num==1:
        return 1
    return fibo(num-1)+fibo(num-2)
def series(num):
    if num==0:
        return 0
    else:
        print(fibo(num))
    return series(num-1)
num=int(input("Enter Number : "))
print(series(num-1))
```

count 0 in given number using recursion

```python
def count_zero(num,c=0):
    if num==0:
        return c
    return count_zero(num//10,c+1) if num%10==0 else count_zero(num//10,c)
num=int(input("Enter Number : "))
print(count_zero(num))
```

check number is Prime using recursion

```python
def check_prime(num,i=2):
    if i==num:
        return print("Prime")
    return print("not Prime") if num%i == 0 else check_prime(num,i+1)

num=int(input("Enter the number : "))
check_prime(num)
```

Reverse the number without extra variable

```python
def reverse_num(num):
    if num<10:
        return num
    return int(str(num%10)+str(reverse_num((num//10))))
num=int(input("Enter Number : "))
print(reverse_num(num))
```
