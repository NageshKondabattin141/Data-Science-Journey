Count Digit without Converting to String

```python
num=int(input("Enter the number : "))
count=0
while num>0:
    count+=1
    num//=10
print(count)
```

Palindrome Number

```python
num=int(input("Enter the number to check Palidrome or not : "))
temp=num
revnum=0
digit=0
if num in range(10):
    print("Yes")
else:
    while num!=0:
        digit=num%10
        revnum=(revnum*10)+digit
        num//=10
        
    if revnum == temp:
        print("Yes")
    else:
        print("No")
        
```

Digit root

```python
num =int(input("Enter Number : "))
digit=0
sumnum=0
if num in range(10):
    print(num)
else:
    while num!=0:
        digit=num%10
        sumnum+=digit
        num//=10
        if num==0:
            num=sumnum
    print(sumnum)
```
