vowels/consonent logic

```python
string = input("Enter a String : ")
vowels="aeiou"
vow_count=0
cons_count=0
for i in string:
    if i in vowels:
        vow_count+=1
    elif i!=" ":
        cons_count+=1
print("Vowels = ",vow_count," Consonent = ",cons_count)
```

Count upper lower special and digits in given string

```python
string= input("Enter String : ")
u=l=s=d=0
for i in string:
    if i.isupper():
        u+=1
    elif i.islower():
        l+=1
    elif i.isdigit():
        d+=1
    else:
        s+=1
print("Uppercase = ",u)
print("Lowercase = ",l)
print("Digits = ",d)
print("Special = ",s)
```

Maximum Repeating Word

```python
string=input("Enter String : ")
l1=string.split()
large=0
unique=[]
for i in l1:
    if l1.count(i)>=large and i not in unique:
        large=l1.count(i)
        unique.append(i)
for i in unique:
    if l1.count(i)==large:
        print(i)
```
