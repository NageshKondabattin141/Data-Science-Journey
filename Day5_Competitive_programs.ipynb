##Tricky warm up

###check number palindrome (no string conversion)

```python
def rev_num(num,rev=0):
    if num == 0:
        return rev
    return rev_num(num//10,rev*10+num%10)

num=int(input("Enter the number"))
if num==rev_num(num):
    print(True)
else:
    print(False)
```

###second largest element (No sorting)

```python
l1=[-10,-20,-30,-28,-98]
first=second=float('-inf')
for i in l1:
    if i>first and i > second:
        second=first
        first=i
    elif i<first and i>second:
        second=i

print(second)
    
```

###Frequency of each element in list

```python
l1=[12,3,12,3,53,56,3,5,6,5,6,7]
unique=[]
for i in l1:
    if i not in unique:
        unique[i]=l1.count(i)
print(unique)
```

###Shift all 0s to end 

```python
l1=[0,22,4,0,6,4,0]
pos=0
for i in range(len(l1)):
    if l1[i]!=0:
        l1[pos],l1[i]=l1[i],l1[pos]
        pos+=1
print(l1)
```

##Brain Test

###Find duplicate in List

```python
def find_duplicate(arr):
    seen = set()

    for num in arr:
        if num in seen:
            return num
        seen.add(num)
l1=[10,30,23,43,53,10]
print(find_duplicate(l1))
    
```

###Find duplicate if arr size 1 to n (use floyd method)

```python
def find_duplicate(num):
    first=num[0]
    next_val=num[0]
    
    while True:
        first=num[first]
        next_val=num[num[first]]
        if first==next_val:
            break
    first=num[0]

    while first != next_val:
        first=num[first]
        next_val=num[next_val]

    return first

l1=[1,3,4,2,4]
print(find_duplicate(l1))
```
