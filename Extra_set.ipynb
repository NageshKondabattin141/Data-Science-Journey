Palindrome Number

```python
num=int(input("Enter the number to check Palidrome or not : "))
temp=num
revnum=0
digit=0
if num in range(10):
    print("Yes")
else:
    while num!=0:
        digit=num%10
        revnum=(revnum*10)+digit
        num//=10
        
    if revnum == temp:
        print("Yes")
    else:
        print("No")
        
```

Second largest element without Sort

```python
l1=[10,8,34,50]
second=0
first=0
for i in l1:
    if i>first :
        second=first
        first=i
    elif i<first and i>second:
        second=i
print(second)
    
```

GCD of two numbers

```python
num1=int(input("Enter number1 : "))
num2=int(input("Enter number2 : "))
maxval=max(num1,num2)
minval=min(num1,num2)
while maxval != 0:
    minval,maxval = maxval,minval % maxval
print("GCD = ",minval)
    
```

Strong Number 

```python
num=int(input("Enter Number : "))
temp = num
sumval=0
while num>0:
    digit=num%10
    fact=1
    for i in range(1,digit+1):
        fact*=i
    sumval+=fact
    num//=10
if temp == sumval:
    print("Strong")
else:
    print("Not Strong")
```

Perfect Number

```python
num=int(input("Enter Number : "))
sumval=0
for i in range(1,num):
    if num%i==0:
        sumval+=i
if num == sumval:
    print("Perfect")
else:
    print("Not Perfect")
```
