```python
num=int(input("Enter Number : "))
count=0
if num==0:
    print("no of digits in number is 1")
else:
    while num!=0:
        num//=10
        count+=1
    print("no of digits in number is ",count)
```

```python
num=int(input("Enter number : "))
revnum=0
lastdigit=0
if num in range(0,10):
    print("reverse of number is ",num)
else:
    while num!=0:
        lastdigit=num%10
        revnum=revnum*10+lastdigit
        num//=10
    print("reverse of number is ",revnum)
```

```python
num = int(input("Enter number to check prime or not : "))
if num==0 or num==1:
    print(num)
else:
    flag=False
    for i in range(2,num):
        if num%i==0:
            print("number is not prime")
            flag=True
            break
    if flag==False:
        print("number is Prime")
```

```python
#prime numbers from given range
start=int(input("Enter starting value : "))
last=int(input("Enter ending value : "))
flag=False
for i in range(start,last+1):
    for j in range(2,i):
        if i%j==0:
            flag=True
            break
    if flag==False:
        print(i)
    flag=False
```

Number patterns

```python
#simple pyramid
num=int(input("Enter Number : "))
for i in range(1,num+1):
    for j in range(1,i+1):
        print(j,end="")
    print()
```

```python
#same number pyramid
num=int(input("Enter Number : "))
for i in range(1,num+1):
    for j in range(1,i+1):
        print(i,end="")
    print()
```

```python
#Reverse pyramid
num=int(input("Enter Number : "))
for i in range(num,0,-1):
    for j in range(i,0,-1):
        print(j,end="")
    print()
```

```python
#same number reverse pyramid
num=int(input("Enter Number : "))
for i in range(num,0,-1):
    for j in range(i,0,-1):
        print(i,end="")
    print()
```

```python
#Centered pyramid
num=int(input("Enter Number : "))
for i in range(1,num+1):
    print(" "*(num-i),end="")
    for j in range(1,i+1):
        print(j,end="")
    for j in range(i-1,0,-1):
        print(j,end="")
    print()
```

```python
#Floyd's pyramid
num=int(input("Enter Number : "))
n=0
for i in range(1,num+1):
    for j in range(1,i+1):
        n+=1
        print(n,end=" ")
    print()
```
